### Criando gráficos com métodos já definidos na biblioteca networkx barabasi_albert_graph
barabasi_albert_graph faz parte da biblioteca NetworkX e é usado para criar um grafo usando o modelo de crescimento preferencial de Barabási-Albert. Esse modelo é uma das maneiras de gerar grafos de escala livre, que possuem distribuição de grau que segue uma lei de potência.

<h4>A assinatura da função barabasi_albert_graph(n, m) recebe dois parâmetros:</h4>

n: é o número de nós (vértices) no grafo que você deseja criar. 


m: é o número de arestas que cada novo nó adicionado ao grafo forma para se conectar aos nós existentes.


No modelo Barabási-Albert, o grafo é construído iterativamente adicionando um novo nó a cada iteração e conectando-o a m nós existentes, com probabilidade proporcional ao grau desses nós. Isso significa que os nós mais conectados têm maior probabilidade de receber novas arestas, refletindo o princípio do "rico fica mais rico" ou "preferência por vínculos a nós já populares".

```python
G = nx.barabasi_albert_graph(100,2)
```

Método draw_spring é frequentemente útil para  visualizar essas redes. O networkx inclui várias opções de visualização. Aqui está uma opção simples.

```python
nx.draw_spring(G)
```

Analisando os graus do grafo

```python
G.degree
```

```python
plt.hist([v for k,v in nx.degree(G)])
```

O método closeness_centrality faz parte da biblioteca NetworkX e é usado para calcular a centralidade de proximidade de cada nó em um grafo. A centralidade de proximidade mede quão próximo um nó está de todos os outros nós do grafo.

A função closeness_centrality(G) recebe como parâmetro um grafo G e retorna um dicionário contendo os valores de centralidade de proximidade para cada nó do grafo.

A centralidade de proximidade de um nó é calculada como a inversa da média das distâncias mais curtas entre esse nó e todos os outros nós do grafo. Quanto menor a média das distâncias mais curtas, maior será a centralidade de proximidade de um nó, indicando que ele está mais próximo de todos os outros nós.

```python
plt.hist(nx.centrality.closeness_centrality(G).values())
```

## Importing from Pandas

Como utilizar o netWorkX quando temos que retirar os dados de um objeto pandas? 

```python
nodes = list(range(100))

df = pd.DataFrame({'from': np.random.choice(nodes, 100),
                   'to': np.random.choice(nodes,100)
                  })
```

```python
df.head(10)
```

```python
G = nx.from_pandas_edgelist(df, source='from', target='to')
```

```python
nx.draw(G)
```

```python
plt.hist([v for k,v in nx.degree(G)])
```
